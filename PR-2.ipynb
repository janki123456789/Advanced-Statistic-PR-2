Definition: 
           Probability is a measure of the likelihood that a particular event will occur.

In simple terms:

Probability tells us how likely something is to happen when the outcome is uncertain.

Probability is used when we cannot predict an outcome with complete certainty but can estimate how often it may occur based on available data.

Formula :

Probability of an Event = Number of favorable outcomes  /  Total number of possible outcomes

Range of Probability Values

- 0 â†’ The event is impossible
- 1 â†’ The event is certain
- Between 0 and 1 â†’ The event may or may not occur


#### (B)

##### 1. Simple Event :

Definition: An event with only one outcome.

Example:
- Event: A student passed the final exam.

            ð¸ = {final_exam_pass = Pass}

This event considers only one condition, so it is a simple event.

##### 2. Compound Event :

Definition: An event that includes two or more simple events.

Types:
- AND event (Intersection)
- OR event (Union)

Example:
- Event: A student studied more than 25 hours and passed the final exam.

       E={study_hours > 25 AND final_exam_pass =Pass}

Since multiple conditions are involved, this is a compound event.

##### 3. Independent Events

Definition: The occurrence of one event does not      affect the occurrence of the other.

Example: 
- Event A: Randomly selecting a student
- Event B: That studentâ€™s attendance percentage value being recorded

The selection of a student does not change the attendance values of other students, so these events are considered independent.

#####  4. Dependent Events

Definition: The occurrence of one event does affect the other.

Example: 
- Event A: A student studied more than 30 hours per week
- Event B: The student passed the final exam

Higher study hours increase the chance of passing, so these events are dependent.

##### 5. Mutually Exclusive Events

Definition: Two events that cannot happen at the same time.

Example:
- Event A: Student passed the final exam
- Event B: Student failed the final exam

A student cannot pass and fail simultaneously, so these events are mutually exclusive.

##### 6. Exhaustive Events

Definition: A complete set of all possible outcomes.

Example:
- Events:
- Student passed the final exam
- Student failed the final exam

Every student must either pass or fail, so these events together are exhaustive.

##### 7. Complementary Events

Definition: One event happens or the opposite happens.

Example:
- Event A: Student passed the final exam
- Complement Aâ€²: Student failed the final exam

           ð‘ƒ(ð´â€²) = 1 âˆ’ ð‘ƒ(ð´)

Passing and failing are complementary events.

##### Contingency Tables:

A contingency table (also called a cross-tabulation or cross-tab) is a table that shows the frequency distribution of variables.
Itâ€™s especially useful for joint, marginal, and conditional probabilities.

```python
import pandas as pd

contingency_table = pd.crosstab(df['group_discussion'], df['final_exam_pass'])
contingency_table
```

```python
total_students = len(df)
total_students
```

Joint Probability :

Joint Probability = P(Participates in group discussion AND Passes exam)

Formula:
      
      P(A AND B) = Number of student who participate   in group discussion AND Pass / Total number of students

```python
joint_count = len(df[(df['group_discussion'] == 'Yes') & (df['final_exam_pass'] == 'Pass')])

joint_probability = joint_count / total_students
joint_probability
```

Marginal Probability:

Marginal Probability = P(Passes Exam)

Formula:

       P(Pass) = Total number of students who pass / Total number of students

```python
pass_count = (df['final_exam_pass'] == 'Pass').sum()

marginal_probability = pass_count / total_students
marginal_probability
```

Conditional Probability :

Conditional Probability = P(Passes exam / Participates in group discussion)

Formula:

     P(Pass/ Group discussion) = P(Pass AND Group discussion) / P(Group discussion)

```python
group_discussion_count = (df['group_discussion'] == 'Yes').sum()

conditional_probability = joint_count / group_discussion_count
conditional_probability
```

```python
print("Joint probability :", joint_probability)
print("Marginal probability :", marginal_probability)
print("Conditional probability :", conditional_probability)
```

Bayesâ€™ Theorem:

Bayesâ€™ Theorem is a way to update probabilities when you have new evidence.
It tells you how likely an event is given that another event has occurred.

Formula:

         P(A|B) = P(B|A) * P(A) / P(B)
Where:
- A = Student passes the exam
- B = Student has hight attendance(>80%)

       P(Pass | Hight Attendance) = P(Hight Attendance | Pass) * P(Pass) / P(Hight Attendance)

```python
P_pass = (df['final_exam_pass'] == 'Pass').sum() / len(df)
P_pass
```

```python
P_high = (df['attendance'] > 80).sum() / len(df)
P_high
```

```python
P_high_given_pass = len(df[(df['final_exam_pass'] == 'Pass') & (df['attendance'] > 80)]) / len(df[df['final_exam_pass'] == 'Pass'])
P_high_given_pass
```

```python
P_pass_given_high = (P_high_given_pass * P_pass) / P_high
P_pass_given_high
```

```python
print("Overall P(Pass):", P_pass)
print("P(Pass | High Attendance):", P_pass_given_high)
```
